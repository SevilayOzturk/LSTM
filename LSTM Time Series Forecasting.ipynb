# ***Converting Forecasting Problem to Supervised Learning Problem***
Long Short-Term Memory (LSTM) is a type of Recurrent Neural Network (RNN) that is particularly well-suited for making predictions on time series data. LSTMs are capable of learning long-term dependencies in data, which makes them ideal for tasks such as forecasting, where the relationship between past and future values is critical.

```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
```

```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

```python
df = pd.read_csv(csv_path)
df
```

```python
df = df[5::6]
df.head()
```

```python
df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
df[:26]
```

```python
temp = df['T (degC)']
temp.plot()
```

[[1,2,3,4,5]] [6]
[[2,3,4,5,6]] [7]
[[3,4,5,6,7]] [8]

x matrix (input) and y vector (output)

```python
def df_to_Matrix(df, window_size = 5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)- window_size):
    row = [[a] for a in df_as_np[i:i+5]]
    X.append(row)
    label = df_as_np[i+5]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X, y = df_to_Matrix(temp, WINDOW_SIZE)

X.shape, y.shape
```

```python
X_train, y_train = X[:60000], y[:60000]
X_val, y_val = X[60000:65000] , y[60000:65000]
X_test, y_test = X[65000:], y[65000:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5,1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
```

```python
cp = ModelCheckpoint('model1/', save_best_only= True)
model1.compile(loss=MeanSquaredError(), optimizer= Adam(learning_rate= 0.0001), metrics = [RootMeanSquaredError()])
```

```python
model1.fit(X_train, y_train, validation_data= (X_val, y_val) , epochs = 10, callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('model1/')
```

```python
train_predictions = model1.predict(X_train).flatten()
train_results = pd.DataFrame(data= {'Train Predictions': train_predictions, 'Actuals': y_train})
train_results
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results['Train Predictions'][:100])
plt.plot(train_results['Actuals'][:100])
```

```python
val_predictions = model1.predict(X_val).flatten()
val_results = pd.DataFrame(data= {'Validation Predictions': val_predictions, 'Actuals': y_val})
val_results
```

```python
plt.plot(val_results['Validation Predictions'][:100])
plt.plot(val_results['Actuals'][:100])
```

```python
test_predictions = model1.predict(X_test).flatten()
test_results = pd.DataFrame(data= {'Test Predictions': test_predictions, 'Actuals': y_test})
test_results
```

```python
plt.plot(test_results['Test Predictions'][:100])
plt.plot(test_results['Actuals'][:100])
```
